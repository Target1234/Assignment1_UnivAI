```python
import re

d1 = dict()
fd = open("data/Hamlet.txt",encoding="utf-8" )
lines = fd.readlines()
fd.close()

cleaned_lines = [line.strip() for line in lines]

list_of_lines_words = [line.split() for line in cleaned_lines]

#Converting all the words into proper format by removing special characters , numbers and making the word lower case for uniformity.
for lines_words in list_of_lines_words:
    for word in lines_words:
        word = re.sub('[^A-Za-z]+', '', word)
        word = word.lower()
        if word in d1:
            d1[word] = d1[word] + 1
        else:
            d1[word] = 1
            
#Sorting the data structure based on frequency and removing the top 30 words.    
topwords = sorted(d1, key=d1.get, reverse=True)
del topwords[:30]

#Set of Hamlet Play words
set1 = set(topwords)
print(len(set1))


```

```python
import numpy as np
import vector

# Computing the Vocabulary by union of different play sets
vocabulary = set1.union(set2, set3, set4, set5, set6, set7)
vocabulary = list(vocabulary)
vocabulary.remove('')

#Printing the Vocabulary
print(vocabulary)

#storing the count of words occured in each play.
vector1 = []
vector2 = []
vector3 = []
vector4 = []
vector5 = []
vector6 = []
vector7 = []
for word in vocabulary:
    if word in d1:
        vector1.append(d1[word])
    else:
        vector1.append(0)
        
for word in vocabulary:
    if word in d2:
        vector2.append(d2[word])
    else:
        vector2.append(0)
        
for word in vocabulary:
    if word in d3:
        vector3.append(d3[word])
    else:
        vector3.append(0) 
        
for word in vocabulary:
    if word in d4:
        vector4.append(d4[word])
    else:
        vector4.append(0)

for word in vocabulary:
    if word in d5:
        vector5.append(d5[word])
    else:
        vector5.append(0)
        
for word in vocabulary:
    if word in d6:
        vector6.append(d6[word])
    else:
        vector6.append(0)

for word in vocabulary:
    if word in d7:
        vector7.append(d7[word])
    else:
        vector7.append(0)
        

#length of vocabulary   
print("Dimensional Space of Vocabulary", len(vocabulary))
```
